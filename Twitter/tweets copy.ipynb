```python
import tweepy as tw       # Importando dados do tweepy
import pandas as pd       # Importar Pandas
```

```python
#Aplicando API key e API Secret

my_api_key = 'gO9Dck4O2qUVVjb1WRRCWdoZZ'
my_api_secret = 'A36i4L3wFVqD5VW6VobBM3W4hqGE0UZP5kvmTXNLlZq6vXx6TR'


#Autenticação

auth = tw.OAuthHandler(my_api_key, my_api_secret)
api= tw.API(auth, wait_on_rate_limit = True)
```

```python
search_query = '[democracia, violencia] - filter:retweets'
```

```python
#Pega tweets da API

tweets = tw.Cursor(api.search_tweets,
                   q=search_query,
                   lang = 'pt',
                   since ='2022-01-01').items(3000)

#Armazena as respostas da API em uma lista
tweet_copy = []

for tweet in tweets:
    tweet_copy.append(tweet)
    


print('Total Tweets coletados: ', len(tweet_copy))
```

```python
# Inicializando o Dataframe
tweets_df = pd.DataFrame()

for tweet in tweet_copy:
    hashtags = []
    
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    
    except:
        pass
    
    tweets_df = tweets_df.append(pd.DataFrame({#'user_name': tweet.user.name, 
                                               'user_location': tweet.user.location,\
                                               'user_description': tweet.user.description,
                                               #'user_verified': tweet.user.verified,
                                               'date': tweet.created_at,
                                               #'tweet':tweet.retweets,
                                               'text': text, 
                                               'hashtags': [hashtags if hashtags else None],
                                               'source': tweet.source}))
    tweets_df = tweets_df.reset_index(drop=True)


# Mostrando o dataframe
tweets_df.head()

```
