## Tratamento Tipo de Dados

```python
df_auto["Preço"] = df_auto["Preço"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Tamanho do motor"] = df_auto["Tamanho do motor"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Comprimento"] = df_auto["Comprimento"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Largura"] = df_auto["Largura"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Altura"] = df_auto["Altura"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Potência (HP)"] = df_auto["Potência (HP)"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
df_auto["Portas"] = df_auto["Portas"].replace(",",".").astype(float)
df_auto["Número de cilindros"] = df_auto["Número de cilindros"].replace("doze",12).replace("oito",8).astype(float)
df_auto["Pico RPM"] = df_auto["Pico RPM"].replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
```
